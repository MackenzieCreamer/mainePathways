```python
# import webdriver
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import re
```

Once you have selenium imported, you need to get the actual library instantiated with a "webdriver" object. Certain websites (such as college board's website) require user input in order to successfully navigate it.

```python
# create webdriver object
driver = webdriver.Firefox()
```

Here, we need go to college board's website for ap courses

```python
driver.get("https://apcourseaudit.inflexion.org/ledger/")
```

Once on college board's website tab for AP courses, we need to highlight only the AP courses we care about, aka those from the state of Maine, with the dropdown on that website for states. The html element for this dropdown has the ID "ledgerState" and we then select the dropdown value for Maine, its statecode, "ME"

```python
select = Select(driver.find_element(By.ID,"ledgerState"))
select.select_by_value('ME')
```

We find the element on the webpage associated with submitting the webpage, so we accomplish that by going to the website, same as we did with the previous cell, and put the ID into our application. Once we've done this, we can submit the page by just submitting a user click response to the element.

```python
element = driver.find_element(By.ID, "search-submit")
element.click()
```

In this section of the code, we have several different pages that we need to get lists of schools from. Once we've gotten the full list of elements on the page that particularly contain links to schools within them, we can save the URLs to visit later and get all the programs for each respective school.

```python
# For loop that iterates 8 times
listOfSchools = []
for x in range(2,9):
    listOfElements = driver.find_element(By.CSS_SELECTOR, ".cb-item-list.cb-item-list-rich.ap-item-list-meta-hidden").find_elements(By.TAG_NAME,"li")
    for i, item in enumerate(listOfElements):
        if(i%2==0):
            hrefLink = item.find_elements(By.TAG_NAME,"a")
            listOfSchools.append(hrefLink[0].get_attribute('href'))
    url = "search.php?p=" + str(x)
    anchorHREF = driver.find_element(By.XPATH, '//a[@href="'+url+'"]')
    anchorHREF.click()
```

Now that we've collected all the college board links that are associated with each of the schools, we can visit those links and get the programs associated with the CollegeBoard affiliated high school. The way in which I added these to the simplifiedSchools spreadsheet that we have shared is by manually finding the school associated with the AP Program school name in our list (some are slightly different in terms of their name/representation, which is why I specify it in this fashion). Anyway, once I've identified the school that was associated with the college board name for it, I manually copy/paste'd the programs into the spreadsheet from the output below the cell once it's fully ran through.

```python
for school in listOfSchools[1:]:
    driver.get(school)
    name = driver.find_element(By.CSS_SELECTOR, ".cb-item-title").get_attribute("innerHTML")
    programs = driver.find_elements(By.CSS_SELECTOR, ".cb-callout")
    schoolPrograms = []
    for program in programs:
        try:
            program = program.find_element(By.TAG_NAME,"a")
            schoolProgram = program.get_attribute("innerHTML")
        except:
            schoolProgram = program.get_attribute("innerHTML")
        returnValue = re.search("[\*<]",schoolProgram)
        try:
            schoolProgram = schoolProgram[:returnValue.span()[0]]
        except:
            pass
        schoolPrograms.append(schoolProgram.strip())
    print(name,end=":")
    for apCourse in schoolPrograms:
        print(apCourse,end=";")
    print("\n\n\n\n---------------------------------------------------------------\n\n\n\n")
```

# Initial College Program Collection Experimentation

I utilized bates college as the program collection test for colleges before moving my work elsewhere once it was decided that others would manually collect this information instead. It follows the same general principles that the above cells use, but for a single college instead of a bunch of high schools. Theoretically you could do something like this for every school, but the time necessary to do this to have it automated wouldn't be worth it, most likely.

```python
driver.get("https://www.bates.edu/academics/")
```

```python
listOfElements = driver.find_element(By.CSS_SELECTOR, ".wp-block-columns.is-layout-flex.wp-container-core-columns-layout-1.wp-block-columns-is-layout-flex").find_elements(By.TAG_NAME,"li")
```

```python
batesList = []
for i, item in enumerate(listOfElements):
    hrefLink = item.find_element(By.TAG_NAME,"a")
    school = (hrefLink.get_attribute('innerHTML'))
    school = school[:school.find("(")]
    school = school.replace("&amp;", "&")
    batesList.append(school.strip())
```

```python
for item in batesList:
    print(item,end=",")
```
